1. Data yang digunakan pada visualisasi ini yaitu data jumlah pencari kerja di setiap kabupaten/kota di Provinsi Lampung

2. Tujuan : Memberikan gambaran jelas terkait distribusi pencari kerja laki-laki dan perempuan serta populasi di setiap kabupaten/kota di Provinsi Lampung

3. User : Mengetahui kompetisi pencari kerja di setiap kabupaten/kota (Masyarakat Umum) dan merumuskan kebijakan tenaga kerja yang lebih efektif seperti penciptaan lapangan kerja, program pelatihan, dll (Pemerintah Daerah)

4. Fungsi : Mengidentifikasi apakah terdapat kesenjangan yang signifikan antara jumlah pencari kerja laki-laki dan perempuan serta pemetaan wilayah yang memiliki pencari kerja terbanyak untuk diberi perhatian lebih dalam hal penyediaan lapangan kerja maupun pelatihan keterampilan

```python
#Menampilkan Data

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('/content/jumlah pencari kerja.csv')
df.head()
```

```python
#Membuat bar plot untuk melihat perbandingan antara jumlah pencari kerja laki-laki dan perempuan di setip kabupaten/kota

fig, ax = plt.subplots(figsize=(12, 6))

width = 0.35

x = np.arange(len(df['Kabupaten/Kota']))

ax.bar(x - width / 2, df['Laki-Laki'], width, label='Laki-Laki')
ax.bar(x + width / 2, df['Perempuan'], width, label='Perempuan')

ax.set_xticks(x)
ax.set_xticklabels(df['Kabupaten/Kota'], rotation=90)

ax.set_xlabel('Kabupaten/Kota')
ax.set_ylabel('Total Pencari Kerja')

ax.set_title('Bar Plot Jumlah Pencari Kerja Laki-Laki dan Perempuan di Setiap Kabupaten/Kota')

ax.legend()

plt.tight_layout()
plt.show()
```

```python
#Membuat line chart untuk melihat perbandingan antara jumlah pencari kerja laki-laki dan perempuan di setip kabupaten/kota

plt.figure(figsize=(12, 6))
plt.plot(df['Kabupaten/Kota'], df['Laki-Laki'], label='Laki-Laki', marker='o')
plt.plot(df['Kabupaten/Kota'], df['Perempuan'], label='Perempuan', marker='o')
plt.xlabel('Kabupaten/Kota')
plt.ylabel('Jumlah Pencari Kerja')
plt.title('Diagram Garis Jumlah Pencari Kerja Laki-Laki dan Perempuan')
plt.xticks(rotation=90)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
#Membuat boxplot untuk melihat apakah terdapat outlier pada data pencari kerja laki-laki dan perempuan dari seluruh kabupaten/kota

plt.figure(figsize=(12, 6))
sns.boxplot(data=df[['Laki-Laki', 'Perempuan']])
plt.title('Boxplot Jumlah Pencari Kerja Laki-laki dan Perempuan')
plt.ylabel('Jumlah Pencari Kerja')
plt.show()
```

```python
#Membuat pie chart untuk melihat proporsi antara jumlah pencari kerja laki-laki dan perempuan

labels = ['Laki-Laki', 'Perempuan']
sizes = [df['Laki-Laki'].sum(), df['Perempuan'].sum()]
colors = ['blue', 'orange']
explode = (0.1, 0)

plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=140)

plt.axis('equal')

plt.title('Proporsi Pencari Kerja Laki-laki dan Perempuan')
plt.show()
```

```python
#Membuat heatmap untuk melihat perbedaan jumlah populasi antar kabupaten/kota dimana semakin gelap warnanya maka semakin banyak penduduk di kabupaten/kota tersebut

plt.figure(figsize=(12, 6))
sns.heatmap(df[['Kabupaten/Kota', 'Laki-Laki', 'Perempuan']].set_index('Kabupaten/Kota'), annot=True, cmap="YlOrRd", fmt="d")
plt.title('Heatmap Jumlah Penduduk per Kabupaten/Kota')
plt.show()
```

```python
#Membuat stacked bar plot untuk melihat total populasi per kabupaten/kota yang mana bar dipecah menjadi 2 bagian (laki-laki dan perempuan) untuk membantu melihat proporsi dari masing-masing jenis kelamin terhadap populasi

plt.figure(figsize=(12, 6))
plt.bar(df['Kabupaten/Kota'], df['Laki-Laki'], label='Laki-Laki')
plt.bar(df['Kabupaten/Kota'], df['Perempuan'], bottom=df['Laki-Laki'], label='Perempuan')

plt.xlabel('Kabupaten/Kota')
plt.ylabel('Total Pencari Kerja')
plt.title('Stacked Bar Plot Total Pencari Kerja per Kabupaten/Kota')
plt.xticks(rotation=90)
plt.legend()
plt.tight_layout()
plt.show()
```
